```python
#1. Write a Pandas program to convert a Panda module Series to Python list and
#it's type.

import pandas as pd
d_series = pd.Series([2,4,5,6,9])
print('panda series and type:')
print(d_series)

print("convert panda series into list:")
print(d_series.tolist())
print(type(d_series.tolist()))
```

```python
#2. Write a Pandas program to add, subtract, multiple and divide two Pandas
#Series.
#Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 9]

import pandas as pd

ss1 = pd.Series([2,4,6,8,10])
ss2 = pd.Series([1,3,5,7,9])

print("add two panda series:")
ssr = ss1 + ss2
print(ssr)

print("subtract two panda series:")
ssr1 = ss1 - ss2
print(ssr1)

print("multiple two panda series:")
ssr2 = ss1 * ss2
print(ssr2)

print("division two panda series:")
ssr3 = ss1 / ss2
print(ssr3)
```

```python
#3. Write a Pandas program to convert Series of lists to one Series.
#Original Series of list
#0 [Red, Green, White]
#1 [Red, Black]
#2 [Yellow]

import pandas as pd

ls = pd.Series([['Red', 'Green','White'],
                ['Red', 'Black'],
                ['Yellow'] ])
print("Printing the Original Series of list")
print(ls)

ls = ls.apply(pd.Series).stack().reset_index(drop = True)
print(ls)
  
```

```python
#4. Write a Pandas program to create the mean and standard deviation of the
#data of a given Series.
#[2, 4, 6, 8, 10, 1, 3, 5, 7, 9]

import pandas as pd

s1 = pd.Series([2, 4, 6, 8, 10, 1, 3, 5, 7, 9])
#print(s1)

print("mean:", s1.mean())
print('standard:', s1.std())
```

```python
#20. Write a Pandas program to find integer index of rows with missing data in a given dataframe.
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'school_code': ['s001','s002','s003','s001','s002','s004'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'date_of_birth': ['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
    'weight': [35, None, 33, 30, 31, None]},
     index = ['t1', 't2', 't3', 't4', 't5', 't6'])
print("Original DataFrame:")
print(df)
index = df['weight'].index[df['weight'].apply(np.isnan)]
df_index = df.index.values.tolist()
print("\nInteger index of rows with missing data in 'weight' column of the said dataframe:")
print([df_index.index(i) for i in index])
```
